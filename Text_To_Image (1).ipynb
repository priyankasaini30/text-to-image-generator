```python
!pip install diffusers transformers accelerate safetensors --quiet

from diffusers import StableDiffusionPipeline
import torch
import numpy as np
import gradio as gr
import os
from huggingface_hub import login
from PIL import Image
from IPython.display import display
import random

from getpass import getpass
os.environ['TOKEN'] = getpass('Enter your Hugging Face token:')



login(token=os.environ['TOKEN'])

pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
device = "cuda" if torch.cuda.is_available() else "cpu"
pipe.to(device)

print("✅ Model loaded successfully!")


def generate_image(prompt, guidance_scale=7.5, seed_input=""):
    if seed_input.strip():
        seed = int(seed_input)
    else:
        seed = random.randint(0, 999999)

    generator = torch.manual_seed(seed)

    with torch.autocast("cuda"):
        result = pipe(
            prompt,
            guidance_scale=guidance_scale,
            num_inference_steps=50,
            generator=generator
        )

    return result.images[0], f"✅ Used seed: {seed}"

prompt = input("Enter your prompt:")
generate_image(prompt)


gr.Interface(
    fn=generate_image,
    inputs=[
        gr.Textbox(label="Prompt", placeholder="e.g., A sunset on Mars"),
        gr.Slider(1, 20, value=7.5, label="Guidance Scale"),
        gr.Textbox(label="Seed (Optional)", placeholder="Leave blank for random")
    ],
    outputs=[
        gr.Image(label="Generated Image"),
        gr.Textbox(label="Info")
    ],
    title="Text-to-Image Generator with Seed"
).launch(share=True)



```
